# Import all images

```python
images = [keras_ocr.tools.read(img) for img in ['1_3L4ThlC_2sKnhMY4fsieXw.jpg.jpg',
                                                'Looking-for-machine-learning-experts-LinkedIn-data-shows-how-to-find-them.png',
                                                'What-is-LinkedIn_-What-is-LinkedIn-Used-For_.jpg']]
```

# Keras_ocr Process

```python
# create pipeline
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
# put all images into the pipeline
prediction_groups = pipeline.recognize(images)
```

```python
# show all images
fig, axs = plt.subplots(nrows=1, ncols=len(images), figsize=(20,20))
axs[0].imshow(images[0])
axs[1].imshow(images[1])
axs[2].imshow(images[2])
```

```python
# with the help of keras_ocr find text from images
fig, axs = plt.subplots(nrows=len(images), figsize=(20,20))
for ax, image, predictions in zip(axs, images, prediction_groups):
  keras_ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```
